```python
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib.pylab import rcParams
import matplotlib.pyplot as plt
```

```python
# Download da série mensal do IBOVESPA
import quandl
data = quandl.get("BCB/7845", trim_start="2009-01-01")
data.head()
```

```python
data.index
```

```python
dt = data.groupby('Date')["Value"].mean().reset_index()
dt.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
(dt.set_index('Date')
   .assign(month=lambda df: df.index.month)
   .groupby('month')['Value'].agg(["mean", "std", "median", "min", "max"])
   .plot(ax=ax, marker="o",linewidth=1))
ax.set_xlabel('Mês')
```

```python
# Renomear a série
data.columns = ['IBOVESPA']
# Visualizar a série temporal
data.plot()
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(data)
plot_pacf(data)
plt.show()
```

```python
ts_log=np.log(data["IBOVESPA"])
ts_log_dif = ts_log - ts_log.shift()
ts_log_dif.dropna(inplace=True)
```

```python
plot_acf(ts_log_dif)
plt.show()
```

```python
plot_pacf(ts_log_dif)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
## ARIMA
model = ARIMA(ts_log,order=(10,1,8))
result_MA = model.fit(disp=-1)
print(result_MA.summary())
```

```python
result_MA.plot_predict(plot_insample=True, alpha=0.05)
plt.show()
```

```python
fig = result_MA.plot_predict(1 , 159)
```

```python
from pmdarima import auto_arima
```

```python
stepwise_model = auto_arima(ts_log, start_p=1, start_q=1,max_p=3, max_q=3, m=12, start_P=0, seasonal=True, d=1, D=1, trace=True, error_action='ignore',
                            suppress_warnings=True, stepwise=True)
```

```python
stepwise_model.summary()
```

```python
print(stepwise_model.aic())
```

```python
stepwise_model.fit(ts_log)
```

```python
future_forecast = stepwise_model.predict(n_periods=10)
```

```python
print(future_forecast)
```

```python
plt.plot(future_forecast)
```

```python
from fbprophet import Prophet
```

```python

X = pd.DataFrame(index=range(0,len(data)))
X['ds'] = data.index
X['y'] = data['IBOVESPA'].values
X.head()
```

```python
m = Prophet(yearly_seasonality=True)
m.fit(X)
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```
