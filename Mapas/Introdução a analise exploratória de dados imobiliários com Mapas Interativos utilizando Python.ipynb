#### Conectando essas informações com mundo da Ciência de Dados

Considerando esse cenário, e conectando essas informações com mundo da Ciência de Dados, vamos introduzir estudo para analise exploratória de dados imobiliários com Mapas Interativos utilizando Python. <br>

Vamos importar um dataset que contém informações de imovéis no estado de São Paulo, a princípio nosso interesse é de apenas utilizar as informações de latitude e longitude para construirmos o mapa. 

```python
import pandas as pd
dados = pd.read_csv("lat_long_imob_SP.csv")
dados.head()
```

```python
import folium
from folium.plugins import MarkerCluster
```

Folium é uma biblioteca que facilita a visualização dos dados que foram manipulados no Python em um mapa interativo de folhetos. Além disso permite a ligação de dados a um mapa para visualizações, além de transmitir visualizações vetoriais / rasterizadas / HTML avançadas como marcadores no mapa.

```python
mapa = folium.Map(location=[-16.1237611, -59.9219642],zoom_start=4)

MarkerCluster = MarkerCluster().add_to(mapa)
```

A função "folium.Map()" criará um objeto fazendo referência onde estará localizado o mapa. Já a "MarkerCluster()" adicionará Clusters ao mapa, ou seja, os pontos geográficos de interesse.

```python
for i in range(len(dados)):
    folium.Marker(
        location=[dados.iloc[i,4], dados.iloc[i,5]], icon=folium.Icon(icon='book').add_to(mapa)
    ).add_to(MarkerCluster)

mapa
```

A função "folium.Marker()" contém diversas informações, é nela que definimos as latitudes e longitudes, etc. <br>
Com auxilio do mapa é possível identificarmos onde estão concentrados imovéis no território de SP, de maneira macro, conseguimos extrair informações pertinentes a geolocalização, etc.
